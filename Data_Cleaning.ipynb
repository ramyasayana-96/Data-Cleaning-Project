```python
# ==========================================================
# STEP 1: Import Libraries
# ==========================================================

import pandas as pd
import numpy as np
from google.colab import files

# ==========================================================
# STEP 2: Upload Dataset
# ==========================================================

print("Please select your Airbnb CSV file")

uploaded = files.upload()

# Get uploaded filename automatically
filename = next(iter(uploaded))

print("Uploaded File:", filename)

# ==========================================================
# STEP 3: Read Dataset
# ==========================================================

df = pd.read_csv(filename, low_memory=False)

print("Dataset Loaded Successfully!")

print(df.head())
```

```python
# ==========================================================
# Dataset Information
# ==========================================================

print("Shape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows:")
display(df.head())

print("\nLast 5 Rows:")
display(df.tail())
```

```python
# ==========================================================
# Before Cleaning Statistics
# ==========================================================

print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nSummary Statistics")
display(df.describe(include='all'))
```

```python
# ==========================================================
# Remove Duplicate Rows
# ==========================================================

before_rows = len(df)

df = df.drop_duplicates()

after_rows = len(df)

print("Rows Before:", before_rows)
print("Rows After :", after_rows)
print("Duplicates Removed:", before_rows - after_rows)
```

```python
# ==========================================================
# Handle Missing Values
# ==========================================================

for col in df.columns:

    if df[col].dtype == "object":

        # Fill missing text values
        df[col] = df[col].fillna("Unknown")

    else:

        # Fill missing numeric values using median
        df[col] = df[col].fillna(df[col].median())

print("Missing values handled successfully.")
```

```python
# ==========================================================
# Fix Data Types
# ==========================================================

for col in ["price", "service fee"]:

    if col in df.columns:

        df[col] = (
            df[col]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
        )

        df[col] = pd.to_numeric(df[col], errors="coerce")

print(df.dtypes)
```

```python
# ==========================================================
# After Cleaning
# ==========================================================

print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nData Types")
print(df.dtypes)
```

```python
# ==========================================================
# Save Cleaned Dataset
# ==========================================================

df.to_csv("Airbnb_Open_Data_Cleaned.csv", index=False)

df.to_excel("Airbnb_Open_Data_Cleaned.xlsx", index=False)

print("Files Saved Successfully!")
```
