```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from google.colab import files

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

sns.set_theme(style="whitegrid")

print("Libraries imported successfully!")


# ============================================================
# 2. UPLOAD DATASET
# ============================================================

print("Please upload your Excel or CSV dataset.")

uploaded = files.upload()

# Get uploaded filename
file_path = list(uploaded.keys())[0]

print("\nFile uploaded successfully!")
print("File name:", file_path)


# ============================================================
# 3. LOAD DATASET
# ============================================================

try:

    # Check file extension
    if file_path.lower().endswith(".xlsx"):

        # Load Excel file
        df = pd.read_excel(file_path)

    elif file_path.lower().endswith(".xls"):

        # Load older Excel file
        df = pd.read_excel(file_path)

    elif file_path.lower().endswith(".csv"):

        # Load CSV file
        df = pd.read_csv(file_path)

    else:

        raise ValueError(
            "Unsupported file format. Please upload CSV or Excel file."
        )

    print("\nDataset loaded successfully!")

except Exception as e:

    print("\nError while loading dataset:")
    print(e)


# ============================================================
# 4. CHECK IF DATASET WAS LOADED
# ============================================================

if "df" in locals():

    print("\nDataset Shape:")
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])

    print("\nFirst 5 Rows:")

    display(df.head())

else:

    print(
        "\nDataset was not loaded. "
        "Please check the uploaded file."
    )


# ============================================================
# 5. CLEAN COLUMN NAMES
# ============================================================

if "df" in locals():

    df.columns = (
        df.columns
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace(r"[^\w]+", "_", regex=True)
    )

    print("\nCleaned Column Names:")

    print(df.columns.tolist())


# ============================================================
# 6. DATASET INFORMATION
# ============================================================

if "df" in locals():

    print("\n")
    print("=" * 70)
    print("DATASET INFORMATION")
    print("=" * 70)

    print("Number of Rows:", df.shape[0])

    print("Number of Columns:", df.shape[1])

    print("\nData Types:")

    display(
        pd.DataFrame({
            "Column": df.columns,
            "Data Type": df.dtypes.astype(str)
        })
    )

    print("\nDataset Info:")

    df.info()


# ============================================================
# 7. CHECK DUPLICATES
# ============================================================

if "df" in locals():

    print("\n")
    print("=" * 70)
    print("DUPLICATE RECORDS")
    print("=" * 70)

    duplicate_count = df.duplicated().sum()

    print(
        "Number of duplicate rows:",
        duplicate_count
    )


# ============================================================
# 8. CHECK MISSING VALUES
# ============================================================

if "df" in locals():

    print("\n")
    print("=" * 70)
    print("MISSING VALUES")
    print("=" * 70)

    missing_table = pd.DataFrame({

        "Column": df.columns,

        "Missing Count":
        df.isnull().sum().values,

        "Missing Percentage":
        (
            df.isnull().sum().values
            / len(df)
            * 100
        )

    })

    display(
        missing_table
        .sort_values(
            by="Missing Count",
            ascending=False
        )
    )


# ============================================================
# 9. NUMERICAL COLUMNS
# ============================================================

if "df" in locals():

    numeric_columns = (
        df.select_dtypes(
            include=np.number
        )
        .columns
        .tolist()
    )

    print("\n")
    print("=" * 70)
    print("NUMERICAL COLUMNS")
    print("=" * 70)

    for column in numeric_columns:

        print("-", column)


# ============================================================
# 10. DESCRIPTIVE STATISTICS
# ============================================================

if "df" in locals() and len(numeric_columns) > 0:

    print("\n")
    print("=" * 70)
    print("DESCRIPTIVE STATISTICS")
    print("=" * 70)

    display(
        df[numeric_columns]
        .describe()
        .T
    )


# ============================================================
# 11. MEAN, MEDIAN, MODE, MIN, MAX, STD
# ============================================================

if "df" in locals() and len(numeric_columns) > 0:

    statistics = pd.DataFrame({

        "Mean":
        df[numeric_columns].mean(),

        "Median":
        df[numeric_columns].median(),

        "Mode":
        df[numeric_columns].mode().iloc[0],

        "Minimum":
        df[numeric_columns].min(),

        "Maximum":
        df[numeric_columns].max(),

        "Standard_Deviation":
        df[numeric_columns].std()

    })

    print("\n")
    print("=" * 70)
    print("DETAILED STATISTICS")
    print("=" * 70)

    display(statistics)


# ============================================================
# 12. CATEGORICAL COLUMNS
# ============================================================

if "df" in locals():

    categorical_columns = (
        df.select_dtypes(
            include=["object", "category"]
        )
        .columns
        .tolist()
    )

    print("\n")
    print("=" * 70)
    print("CATEGORICAL COLUMNS")
    print("=" * 70)

    for column in categorical_columns:

        print(
            column,
            "->",
            df[column].nunique(),
            "unique values"
        )


# ============================================================
# 13. CATEGORICAL DISTRIBUTION
# ============================================================

if "df" in locals():

    print("\n")
    print("=" * 70)
    print("CATEGORICAL DISTRIBUTIONS")
    print("=" * 70)

    for column in categorical_columns:

        print("\nColumn:", column)

        display(
            df[column]
            .value_counts()
            .head(10)
        )


# ============================================================
# VISUALIZATION 1
# HISTOGRAM
# ============================================================

if "df" in locals() and len(numeric_columns) >= 1:

    column = numeric_columns[0]

    plt.figure(figsize=(10, 6))

    sns.histplot(
        data=df,
        x=column,
        bins=30,
        kde=True
    )

    plt.title(
        f"Distribution of {column}"
    )

    plt.xlabel(column)

    plt.ylabel("Frequency")

    plt.tight_layout()

    plt.show()

    print("\nINSIGHT:")

    print(
        f"The histogram shows the distribution of {column}. "
        f"It helps identify the most common range of values "
        f"and whether the data is skewed."
    )


# ============================================================
# VISUALIZATION 2
# BOX PLOT
# ============================================================

if "df" in locals() and len(numeric_columns) >= 1:

    column = numeric_columns[0]

    plt.figure(figsize=(10, 5))

    sns.boxplot(
        x=df[column]
    )

    plt.title(
        f"Box Plot of {column}"
    )

    plt.xlabel(column)

    plt.tight_layout()

    plt.show()

    print("\nINSIGHT:")

    print(
        f"The box plot shows the median, spread, "
        f"and potential outliers in {column}."
    )


# ============================================================
# VISUALIZATION 3
# BAR CHART
# ============================================================

if "df" in locals() and len(categorical_columns) >= 1:

    column = categorical_columns[0]

    top_categories = (
        df[column]
        .value_counts()
        .head(10)
    )

    plt.figure(figsize=(10, 6))

    sns.barplot(
        x=top_categories.values,
        y=top_categories.index
    )

    plt.title(
        f"Top 10 Categories in {column}"
    )

    plt.xlabel("Number of Records")

    plt.ylabel(column)

    plt.tight_layout()

    plt.show()

    print("\nINSIGHT:")

    print(
        f"The bar chart shows the most common categories "
        f"in {column}. The largest category is the dominant "
        f"category in the dataset."
    )


# ============================================================
# VISUALIZATION 4
# SCATTER PLOT
# ============================================================

if "df" in locals() and len(numeric_columns) >= 2:

    x_column = numeric_columns[0]

    y_column = numeric_columns[1]

    plt.figure(figsize=(10, 6))

    sns.scatterplot(
        data=df,
        x=x_column,
        y=y_column,
        alpha=0.5
    )

    plt.title(
        f"{x_column} vs {y_column}"
    )

    plt.xlabel(x_column)

    plt.ylabel(y_column)

    plt.tight_layout()

    plt.show()

    correlation = df[
        [x_column, y_column]
    ].corr().iloc[0, 1]

    print("\nCorrelation:", round(correlation, 3))

    print("\nINSIGHT:")

    print(
        f"The scatter plot shows the relationship between "
        f"{x_column} and {y_column}. "
        f"The correlation value is {round(correlation, 3)}."
    )


# ============================================================
# VISUALIZATION 5
# CORRELATION HEATMAP
# ============================================================

if "df" in locals() and len(numeric_columns) >= 2:

    correlation_matrix = df[
        numeric_columns
    ].corr()

    plt.figure(figsize=(12, 8))

    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap="coolwarm",
        fmt=".2f",
        linewidths=0.5
    )

    plt.title(
        "Correlation Heatmap"
    )

    plt.tight_layout()

    plt.show()

    print("\nINSIGHT:")

    print(
        "The correlation heatmap shows the strength "
        "and direction of relationships between numerical "
        "variables."
    )


# ============================================================
# VISUALIZATION 6
# SECOND CATEGORICAL VARIABLE
# ============================================================

if "df" in locals() and len(categorical_columns) >= 2:

    column = categorical_columns[1]

    top_categories = (
        df[column]
        .value_counts()
        .head(10)
        .index
    )

    filtered_df = df[
        df[column].isin(top_categories)
    ]

    plt.figure(figsize=(10, 6))

    sns.countplot(
        data=filtered_df,
        y=column,
        order=top_categories
    )

    plt.title(
        f"Distribution of {column}"
    )

    plt.xlabel("Count")

    plt.ylabel(column)

    plt.tight_layout()

    plt.show()

    print("\nINSIGHT:")

    print(
        f"The chart shows the distribution of the most "
        f"common categories in {column}."
    )


# ============================================================
# VISUALIZATION 7
# PAIRPLOT
# ============================================================

if "df" in locals() and len(numeric_columns) >= 2:

    selected_columns = numeric_columns[:5]

    pairplot_data = df[
        selected_columns
    ].dropna()

    sns.pairplot(
        pairplot_data
    )

    plt.show()

    print("\nINSIGHT:")

    print(
        "The pair plot provides an overview of relationships "
        "between multiple numerical variables and can reveal "
        "correlations, clusters, and outliers."
    )


# ============================================================
# 14. SAVE STATISTICS
# ============================================================

if "statistics" in locals():

    statistics.to_csv(
        "descriptive_statistics.csv"
    )

    print(
        "\nDescriptive statistics saved successfully."
    )


# ============================================================
# 15. SAVE CORRELATION MATRIX
# ============================================================

if "correlation_matrix" in locals():

    correlation_matrix.to_csv(
        "correlation_matrix.csv"
    )

    print(
        "Correlation matrix saved successfully."
    )


# ============================================================
# 16. FINAL SUMMARY
# ============================================================

if "df" in locals():

    print("\n")
    print("=" * 70)
    print("EDA COMPLETED SUCCESSFULLY!")
    print("=" * 70)

    print(
        "Total Rows:",
        df.shape[0]
    )

    print(
        "Total Columns:",
        df.shape[1]
    )

    print(
        "Numerical Columns:",
        len(numeric_columns)
    )

    print(
        "Categorical Columns:",
        len(categorical_columns)
    )

    print(
        "Duplicate Rows:",
        df.duplicated().sum()
    )

    print(
        "Total Missing Values:",
        df.isnull().sum().sum()
    )
```
